## 3. Correlation and Visualization

#### Compute correlation matrix


```python
correlation_matrix = df.corr(numeric_only=True)
```

#### Display correlation matrix


```python
correlation_matrix
```



#### Heatmap of correlation matrix


```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

-----------------------------------------------
## 4. Grouping and Aggregation

#### Average population by continent


```python
df.groupby('Continent').mean(numeric_only=True)
```



#### Filter rows where continent contains 'Oceania'


```python
df[df['Continent'].str.contains('Oceania')]
```


#### Average population by continent, sorted by 2022 Population


```python
df.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
```

-----------------------------------------------
## 5. Creating a Subset for Historical Population

#### Selecting relevant columns for population trends


```python
df_dummy = df[['Continent', '1970 Population', '1980 Population', '1990 Population', 
               '2000 Population', '2010 Population', '2015 Population', 
               '2020 Population', '2022 Population']]
```



#### Group by continent and compute mean


```python
df2 = df_dummy.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
```


#### Transpose the dataframe for better visualization


```python
df3 = df2.transpose()
df3
```

#### Line plot for population trends by continent


```python
df3.plot(figsize=(10, 6), title="Population Growth by Continent")
```
